```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
train_data = pd.read_csv("/train_data.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/test_data.csv")
test_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

#Saved all Survived from train_data
y = train_data["Survived"]

#Select four column as consider the feature for train and prediction

features = ["Pclass", "Sex", "SibSp", "Parch", "Embarked"]
#Store the train_data selected features
X = pd.get_dummies(train_data[features])

#Store the test_data selected features
X_test = pd.get_dummies(test_data[features])

#use the sklearn -> RandomForestClassifier model with epoch, depth, and random state

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
#Train the model with train_data feature and survived people from train data
model.fit(X, y)
#Now time to do perdiction from test data as per features who survived
predictions = model.predict(X_test)
#Store the ouput as DataFrame for test data as per prediction for Survived
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
# Store the output in submission file
output.to_csv('submission.csv', index=False)
print(output)
```
